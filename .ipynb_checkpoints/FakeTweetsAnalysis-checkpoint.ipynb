## 2. Data Cleaning
- lowercase (done)
- remove punctuations, numbers (done)
- remove unwanted words (urls, html, stopwords) (done)
- stemming words
- word cloud

```python
def clean_text(text):
    text = text.lower()
    # url
    text = re.sub(r'https?:\/\/t.co\/[A-Za-z0-9]+', '', text)
    # html
    text = re.sub(r'<.*?>', '', text)
    # punctuation, numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # stopwords
    text = ' '.join([word for word in text.split() if not word in set(stopwords.words('english'))])
    return text


train['cleaned_text'] = train['text'].apply(lambda tweet: clean_text(tweet))
test['cleaned_text'] = test['text'].apply(lambda tweet: clean_text(tweet))
test.head()
```

## 3. Modeling

```python
X, y = train['cleaned_text'], train['target']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=40)
```

### 3.1 GloVe for Vectorization
GloVe dataset from: https://www.kaggle.com/rtatman/glove-global-vectors-for-word-representation
<br>GloVe: an unsupervised learning algorithm for obtaining vector representations for words.

```python
embedding_index = {}
with open('glove.6B.100d.txt','r') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vectors = np.asarray(values[1:], dtype='float32')
        embedding_index[word] = vectors
f.close()
```

### 3.2 Word Embedding

```python
MAX_LEN = 50
EMBEDDING_DIM = 100

tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)

train_sequences = tokenizer.texts_to_sequences(X_train)
train_padded = pad_sequences(train_sequences, maxlen=MAX_LEN, truncating='post', padding='post')
train_word_index = tokenizer.word_index

val_sequences = tokenizer.texts_to_sequences(X_val)
val_padded = pad_sequences(val_sequences, maxlen=MAX_LEN, truncating='post', padding='post')
val_word_index = tokenizer.word_index
```

```python
num_words = len(train_word_index) + 1
embedding_matrix = np.zeros((num_words, EMBEDDING_DIM))

for word, i in train_word_index.items():
    embedding_vec = embedding_index.get(word)
    if embedding_vec is not None:
        embedding_matrix[i] = embedding_vec
```

### 3.3 LSTM (Long short-term memory) Networks

```python
model = tf.keras.Sequential([
    
    tf.keras.layers.Embedding(num_words, EMBEDDING_DIM, input_length=MAX_LEN),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy',tf.keras.metrics.AUC()])
model.summary()
```

```python
history = model.fit(train_padded, y_train, batch_size=4, epochs=3, validation_data=(val_padded, y_val))
```

```python
# submission to Kaggle
test_sequences = tokenizer.texts_to_sequences(test['cleaned_text'])
test_padded = pad_sequences(test_sequences, maxlen=MAX_LEN, truncating='post', padding='post')
pred = model.predict(test_padded)
pred = np.round(pred).astype(int)
pred_df = pd.DataFrame({'id': test['id'].values.tolist(), 'target': pred.reshape(pred.shape[0])})
pred_df.to_csv('submission.csv', index=False)
```
